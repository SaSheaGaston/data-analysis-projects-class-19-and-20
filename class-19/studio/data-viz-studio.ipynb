```python
# Here is the setup of the dataframe. Feel free to use .info() or .head() to 
## get a better understanding of what is inside the dataframe!

import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
from matplotlib import style

df = pd.read_csv(r"C:\Users\Sa Shea\data-analysis-projects-SaSheaG\data-analysis-projects-class-19-and-20\class-19\studio\books.csv", on_bad_lines='skip')

years = []

for y in df['publication_date']:
    date = y.split("/")
    years.append(date[2])

df["year"] = years

df.head()
```

```python

years = list(df["year"].unique())
years
```

```python
len(years)
```

```python
years_int = [int(i) for i in years]
years_int
```

```python
sorted_years_int = sorted(years_int)
sorted_years_int
```

```python
df["year"] = df["year"].astype(int)
```

```python
# Use your Python skills to create a visualization here.

# Number of books published per year




books_published = []


for i in sorted_years_int:
	book_count = df[df["year"] == i]["bookID"].count()
	books_published.append(book_count)

plt.bar(x=sorted_years_int, height=books_published,width=.4, color ='maroon')

plt.title("Books Published per Year")
plt.xlabel("Year")
plt.ylabel("Total Books")
```

```python
# Use your Python skills to create a visualization here.

# Number of books published by each publisher

df["publisher"].unique()
```

```python
publishers = list(df["publisher"].unique())
len(publishers)
```

```python
publisher_counts = df.groupby("publisher")["bookID"].count()

plt.hist(publisher_counts, bins=10)
plt.xlabel("Books Published")
plt.ylabel("Number of Publishers")
plt.title("Distribution of Books Published per Publisher")
```

```python
top_publishers = (df.groupby("publisher")["bookID"].count().sort_values(ascending=False).head(20))
plt.figure(figsize=(10, 6))
plt.barh(top_publishers.index, top_publishers.values)
plt.gca().invert_yaxis()
plt.title("Top 20 Publishers by Number of Books Published")
plt.xlabel("Number of Books")
plt.ylabel("Publisher")
```
